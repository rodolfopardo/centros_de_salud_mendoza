```python
#Importamos las librerias necesarias para operar

import folium
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from folium.plugins import HeatMap
```

```python
#Importando el archivo

centros_de_salud = pd.read_csv('/Users/rodolfopardo/Downloads/georreferenciacion-centros-salud.csv', sep = ";")
```

```python
#Investigando nuestra data

centros_de_salud.head()
```

```python
#Investigamos los ultimos valores

centros_de_salud.tail()
```

```python
#Descubrimos el formato de nuestro dataframe

centros_de_salud.shape
```

<font color=green> Tenemos 325 centros de salud con 8 variables a analizar

```python
#Tipos de datos del dataframe 

centros_de_salud.dtypes
```

```python
#Nombre de columnas ideales

columnas = ['id', 'id_dep', 'departamento', 'latitud', 'longitud', 'nombre', 'domicilio', 'tipo']
```

```python
#Cambiamos el nombre de las variables 

centros_de_salud.columns = columnas
```

```python
#Las nuevas columnas 

centros_de_salud.columns
```

```python
#Buscando valores perdidos de la data

centros_de_salud.isna().sum()
centros_de_salud = centros_de_salud.dropna()
```

```python
#Comenzamos a agrupar la data
#Departamentos con mas centros de salud 

centros_de_salud.groupby('departamento')['id'].count().sort_values(ascending = False)
```

<font color=blue>San Rafael y Maipú son los departamentos con más centros de salud en la provincia de Mendoza

```python
#Tipos de centros de salud distribuidos en la provincia de Mendoza

centros_de_salud.groupby('tipo')['id'].count().sort_values(ascending = False)
```

<font color=red>Los 365 datos de nuestra base son precisamente centros de salud. 

```python
#Vamos a generar Pivot Tables para guardar en nuestra base 

centros_por_dpt = pd.pivot_table(centros_de_salud, index = 'departamento',
                                values = 'id',
                                aggfunc = ['count'])
```

```python
centros_por_dpt.to_csv('centros_por_departamento.csv')
```

## Representacion en mapas 

```python
mdz_white=folium.Map(location=[-34.1931714, -69.2631543],
                tiles='stamentoner', zoom_start=7)

mdz_white.save('mendoza_white.html')
```

```python
display(mdz_white)
```

```python
mdz_geo=folium.Map(location=[-34.1931714, -69.2631543],
                tiles='openstreetmap', zoom_start=7)
```

```python
display(mdz_geo)
mdz_geo.save('mendoza_geo.html')
```

```python
centros_de_salud['latitud'] = centros_de_salud['latitud'].str[:-4]
```

```python
centros_de_salud['longitud'] = centros_de_salud['longitud'].str[:-4]
```

```python
centros_de_salud['longitud']
```

```python
#Ya tenemos los mapas, ahora es turno de meter las geolocalizaciones indicadas 
#Armamos nuestras variables con latitud y longitud


latitud = list(centros_de_salud['latitud'])
longitud = list(centros_de_salud['longitud'])
```

```python
for i in range(len(latitud)):
    folium.Marker([float(latitud[i]), float(longitud[i])], popup='Centros de Salud', 
                   icon=folium.Icon(color='blue')).add_to(mdz_geo)

    
```

```python
for i in range(len(latitud)):
    folium.Marker([float(latitud[i]), float(longitud[i])],popup='Centros de Salud',
                   icon=folium.Icon(color='red')).add_to(mdz_white)
```

```python
heat_df = centros_de_salud[['latitud', 'longitud']]
```

```python
heat_data = [[row['latitud'],row['longitud']] for index, row in heat_df.iterrows()]
```

```python
map_hooray = folium.Map(location=[-34.1931714, -69.2631543],
                    zoom_start =7)
HeatMap(heat_data).add_to(map_hooray)
```
